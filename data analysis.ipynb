```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df5 = pd.read_csv("out2/5py.csv")
df6 = pd.read_csv("out2/6py.csv")
df7 = pd.read_csv("out2/7py.csv")
```

```python
df5_train = df5["train_epoch_acc"]
df5_valid = df5["valid_epoch_acc"]
df6_train = df6["train_epoch_acc"]
df6_valid = df6["valid_epoch_acc"]
df7_train = df7["train_epoch_acc"]
df7_valid = df7["valid_epoch_acc"]
```

```python
fig, ax = plt.subplots(figsize=(5,4))

ax.plot(list(range(1,11)), df5_train, label="5.py train. acc.", c="black", marker=".", linestyle="dashed")
ax.plot(list(range(1,11)), df5_valid, label="5.py valid. acc.", c="black", marker="x", linestyle="dashdot")
#plt.plot(df6_train, label="6.py train. acc.")
#plt.plot(df6_valid, label="6.py valid. acc.")
#plt.plot(df7_train, label="7.py train. acc.")
#plt.plot(df7_valid, label="7.py valid. acc.")

ax.legend()
ax.grid()

ax.set_xlabel("Epoch")
ax.set_ylabel("Accuracy (%)")
```

```python
fig, axs = plt.subplots(1,3, figsize=(6,2))

axs[0].plot(list(range(1,11)), df5_train, label="Train. acc.", c="black", marker=".", linestyle="dashed")
axs[0].plot(list(range(1,11)), df5_valid, label="Valid. acc.", c="black", marker="x", linestyle="dashdot")
axs[1].plot(list(range(1,11)), df6_train, label="Train. acc.", c="black", marker=".", linestyle="dashed")
axs[1].plot(list(range(1,11)), df6_valid, label="Valid. acc.", c="black", marker="x", linestyle="dashdot")
axs[2].plot(list(range(1,11)), df7_train, label="Train. acc.", c="black", marker=".", linestyle="dashed")
axs[2].plot(list(range(1,11)), df7_valid, label="Valid. acc.", c="black", marker="x", linestyle="dashdot")

axs[0].set_title("a) 5.py")
axs[1].set_title("b) 6.py")
axs[2].set_title("c) 7.py")

for ax in axs:
    ax.grid()
    ax.set_ylim(0,105)

    ax.set_xlabel("Epoch")
    ax.set_ylabel("Accuracy (%)")
    
lines_labels = [axs[0].get_legend_handles_labels()]
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
fig.legend(lines, labels, ncol=2, loc="upper center", bbox_to_anchor=(0.5, 0.05))
plt.tight_layout()
```
